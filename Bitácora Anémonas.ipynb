#### Introducción

Las anémonas pertenecen al Phylum Cnidaria, en el cual también se encuentran las medusas, los corales y las hidras, este grupo se caracteriza por tener células urticantes llamadas cnidocitos, los cuales segregan toxinas, como las actinoporinas que actúan haciendo poros en membrana plasmática, citolisinas que actúan en la célula causando lisis (ruptura de membrana), o destrucción de eritrocitos hemolisinas.

Para la identificación de especies de anémonas no basta con la clasificación morfológica, ya que la identificación de la mayoría de los hexacorales de cuerpo blando requiere exámenes morfológicos e histológicos de especímenes preservados y experiencia para la correcta interpretación de las características observadas (Haussermann 2004).

Por lo que son necesarios estudios:  
* Morfofisiológicos
* Histológicos
* Moleculares  

Para identificación de especies de anémonas.  
Las células urticantes presentes en anémonas forman el cnidoma, que es la distribución, medidas y tipos de cnidocitos que presenta una especie determinada, por cada región del cuerpo. La caracterización del cnidoma es extremadamente relevante para la identificación taxonómica de anémonas, Por lo que la técnica tisular constituye una herramienta diagnóstica de gran relevancia, porque se pueden reconocer los diferentes tipos de cnidocitos

#### Metodo

* Se realizo la colecta en Tecolutla Veracruz, trayendo 10 organismos, estos fueron aclimatados y mantenidos en el PIMVS-CIDMIRA del laboratorio de Ecotoxicología de la UAM-I. 
* Se llevo a cabo un registro de parámetros morfométricos y fisicoquímicos del acuario donde se encuentran. 
* osteriormente se seleccionó a dos anémonas para anestesiarlas, para esto se realizaron dos métodos; Uno con Sales de Mg y cristales de mentol, y el otro reduciendo la temperatura gradualmente.
* Tanto el número de coronas, tentáculos e hileras de verrugas fueron contados uno por uno con un microscopio de disección.  
* Se realizaron cortes de; tentáculos, región oral, columna, gastrodermis y discopedio de ambas anémonas, para análisis histológico y molecular 

##### Para análisis histológico:

* Se fijaron en solución Bouin y formol al 10%
* Las secciones fueron procesadas mediante técnica histológica obteniendo cortes de 5 µm usando tinción H-E y H-F analizadas con microscopio óptico, posteriormente se analizó cada laminilla específicamente para la caracterización del cnidoma, cabe destacar que para el análisis de cnidoma sólo se ocuparon 14 laminillas de una sola anémona.
* Se enfocó con objetivo 40x para posteriormente pasar a 100x 
* Cada laminilla fue analizada en zic zac con objetivo 100x 
* Se realizo el conteo de cnidocitos de cada campo observado (n=435) a 100x, sin especificar el tipo.
* Se tomaron fotografías de cada preparación y fueron procesadas en dos programas: Lightroom y Zerene Stacker. 
* De cada tipo de cnidocisto fueron elegidas al azar, 30 cápsulas sin des-cargar y se midió en cada una el largo y el ancho mediante con el programa AxioVision 4.8, se registró, además la frecuencia de aparición de cada tipo y se calcularon los valores de media, desviación estándar y rango de cada medida obtenida.

##### Para análisis molecular:

La extracción de ADN se realizó empleando el Kit de PROMEGA con modificaciones del protocolo original propuesto por el fabricante. 
Para llevar a cabo el análisis molecular se utilizó material conservado en un ultracongelador (REVCO) -80°C. Posteriormente se realizó la cuantificación del ADN extraído en el equipo NanoDrop 2000. La amplificación vía PCR de los genes 16S y COI se realizó según lo descrito por Palumbi et al (2002) y Folmer et al (1994), respectivamente, con el kit de amplificación de PROMEGA. Los primers y condiciones del ciclo de PCR que se utilizaron se muestran en la Tabla número 1 y en la tabla 2 los reactivos que se ocuparon.

*  Anestesia

La anestesia reportada, usando cristales de mentol y sales de magnesio (Mg) no resulto favorable ya que en menos de 20 min el organismo comenzó a irritarse, desprendiendo un mucus en todo su cuerpo, se probo con la anestesia reduciendo la temperatura gradualmente, llegando hasta los 0° C y así se logró mantener al organismo en completa relajación, para posteriormente proceder con los cortes.
